```python
# ── DATASET ──────────────────────────────────────────────────
# 20 popular products across 5 categories
# Real-world inspired prices (INR) — April 2026
```

```python
import numpy as np
```

```python
products = np.array([
    "boAt Airdopes 141", "JBL Flip 6", "Sony WH-1000XM5",
    "Samsung Galaxy Buds2", "Noise ColorFit Pro 4",
    "Xiaomi 14 Ultra", "Samsung S24 FE", "OnePlus 12R",
    "Realme GT 6T", "iQOO Z9s Pro",
    "HP Victus 15", "Lenovo IdeaPad Slim 3", "ASUS VivoBook 15",
    "Dell Inspiron 15", "Acer Aspire Lite",
    "Pigeon Healthifyme Mixer", "Prestige Iris 750W",
    "Philips HL7756", "Bajaj Rex 500W", "Butterfly Matchless Plus"
])
```

```python
categories = np.array([
    "Audio", "Audio", "Audio", "Audio", "Audio",
    "Mobile", "Mobile", "Mobile", "Mobile", "Mobile",
    "Laptop", "Laptop", "Laptop", "Laptop", "Laptop",
    "Appliance", "Appliance", "Appliance", "Appliance", "Appliance"
])
```

```python
flipkart_prices = np.array([
    1299, 9499, 24999, 7999, 2199,
    89999, 54999, 39999, 27999, 22999,
    67990, 41990, 38990, 57990, 28990,
    2199, 3499, 3199, 1899, 2699
])
```

```python
amazon_prices = np.array([
    1199, 9999, 26499, 8499, 2099,
    91999, 52999, 41999, 26999, 23499,
    69990, 39990, 37490, 59990, 27990,
    2099, 3299, 3399, 1799, 2599
])
```

# ── ANALYSIS 1: Price Differences ────────────────────────────

```python
print("=" * 60)
print("📊 PRICE DIFFERENCE ANALYSIS (Flipkart - Amazon, INR)")
print("=" * 60)

diff = flipkart_prices - amazon_prices
pct_diff = ((flipkart_prices - amazon_prices) / amazon_prices) * 100

for i in range(len(products)):
    platform = "✅ Amazon cheaper" if diff[i] > 0 else "✅ Flipkart cheaper" if diff[i] < 0 else "🟰 Same price"
    print(f"{products[i]:<30} | Diff: ₹{diff[i]:>+6} | {platform}")
```

# ── ANALYSIS 2: Category-wise Winner ────────────────────────

```python
print("\n" + "=" * 60)
print("🏆 CATEGORY-WISE PLATFORM WINNER")
print("=" * 60)

unique_cats = ["Audio", "Mobile", "Laptop", "Appliance"]
cat_savings = {}

for cat in unique_cats:
    mask = categories == cat
    fk_avg = np.mean(flipkart_prices[mask])
    az_avg = np.mean(amazon_prices[mask])
    winner = "Flipkart" if fk_avg < az_avg else "Amazon"
    savings = abs(fk_avg - az_avg)
    cat_savings[cat] = savings
    print(f"{cat:<12} | FK Avg: ₹{fk_avg:>8.0f} | AZ Avg: ₹{az_avg:>8.0f} | 🏆 {winner} saves ₹{savings:.0f}/product")
```

# ── ANALYSIS 3: Statistical Deep-Dive ────────────────────────

```python
print("\n" + "=" * 60)
print("📈 STATISTICAL SUMMARY")
print("=" * 60)

print(f"\n{'Metric':<25} {'Flipkart':>12} {'Amazon':>12}")
print("-" * 50)
print(f"{'Mean price (₹)':<25} {np.mean(flipkart_prices):>12,.0f} {np.mean(amazon_prices):>12,.0f}")
print(f"{'Median price (₹)':<25} {np.median(flipkart_prices):>12,.0f} {np.median(amazon_prices):>12,.0f}")
print(f"{'Std deviation (₹)':<25} {np.std(flipkart_prices):>12,.0f} {np.std(amazon_prices):>12,.0f}")
print(f"{'Min price (₹)':<25} {np.min(flipkart_prices):>12,} {np.min(amazon_prices):>12,}")
print(f"{'Max price (₹)':<25} {np.max(flipkart_prices):>12,} {np.max(amazon_prices):>12,}")
```

# ── ANALYSIS 4: Smart Buyer Recommendation ───────────────────

```python
print("\n" + "=" * 60)
print("💡 SMART BUYER ALGORITHM — WHERE TO BUY EACH PRODUCT")
print("=" * 60)

best_platform = np.where(flipkart_prices <= amazon_prices, "Flipkart 🛒", "Amazon 📦")
potential_savings = np.abs(diff)
total_if_all_flipkart = np.sum(flipkart_prices)
total_if_all_amazon   = np.sum(amazon_prices)
total_smart_shopping  = np.sum(np.minimum(flipkart_prices, amazon_prices))

for i in range(len(products)):
    print(f"{products[i]:<30} → {best_platform[i]}  | Save ₹{potential_savings[i]:,}")

print(f"\n💰 If you bought ALL on Flipkart:      ₹{total_if_all_flipkart:,}")
print(f"💰 If you bought ALL on Amazon:        ₹{total_if_all_amazon:,}")
print(f"🏆 Smart Shopping (best of both):      ₹{total_smart_shopping:,}")
print(f"💚 Total savings with smart shopping:  ₹{total_if_all_flipkart - total_smart_shopping:,} vs Flipkart")
print(f"💚 Total savings with smart shopping:  ₹{total_if_all_amazon - total_smart_shopping:,} vs Amazon")
```

# ── ANALYSIS 5: Discount Detection ───────────────────────────

```python
print("\n" + "=" * 60)
print("🔥 BIG DEAL DETECTOR (>3% cheaper than competitor)")
print("=" * 60)

big_deals_fk = pct_diff > 3    # Amazon is 3%+ cheaper than Flipkart
big_deals_az = pct_diff < -3   # Flipkart is 3%+ cheaper than Amazon

print("\n🛒 Best deals on AMAZON (>3% cheaper):")
for i in np.where(big_deals_fk)[0]:
    print(f"  → {products[i]:<30} | Amazon ₹{amazon_prices[i]:,} vs Flipkart ₹{flipkart_prices[i]:,} | {abs(pct_diff[i]):.1f}% cheaper")

print("\n📦 Best deals on FLIPKART (>3% cheaper):")
for i in np.where(big_deals_az)[0]:
    print(f"  → {products[i]:<30} | Flipkart ₹{flipkart_prices[i]:,} vs Amazon ₹{amazon_prices[i]:,} | {abs(pct_diff[i]):.1f}% cheaper")
```
